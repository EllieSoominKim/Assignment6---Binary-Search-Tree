## #02

1. [n/2]개  
풀이: 완전 이진 트리에서 마지막 레벨을 제외한 나머지는 포화 이진 트리이고, 노드들은 왼쪽부터 채워진다.
임의의 노드 i에 대해 2i와 2i+1이 자식 노드이므로, 자식을 가지는 내부 노드는 인덱스 1 ~ ⌊n/2⌋ 에 해당하는 노드들이다.
따라서 내부 노드 수 = ⌊n/2⌋, 리프 노드 수 = n - ⌊n/2⌋ = ⌈n/2⌉ 개
2. 리프노드 수 = 내부 노드 수 +1(n이 홀수일 때), 또는 같음(n이 짝수일 때)  
즉, 리프노드 수 - 내부 노드 수 = 0 또는 1

## #03
1.  전위 순회: 5-3-1-8-7-6-9  
풀이: 전위순회는 root->left->right 순서
2.  중위 순회: 1-3-5-6-7-8-9  
풀이: 중위순회는 left->root->right 순서
3.  후위 순회: 1-3-6-7-9-8-5  
풀이: 후위순회는 left->right->root 순서
4. Theorem: 이진 검색 트리 T의 중위 순회는 노드들을 정렬된 순서로 방문한다.  

Proof: 트리의 높이 h에 대한 수학적 귀납법으로 증명한다.  

Base case (h = 0): 트리가 비어있는 경우(노드 0개), 방문하는 노드가 없으므로 자명하게 정렬된 순서다.  

Base case (h = 1): 트리가 단 하나의 노드(루트)만 갖는 경우, 유일한 노드 하나를 방문하는 것은 자명하게 정렬된 순서다.  

Inductive hypothesis: h보다 작은 높이를 갖는 모든 이진 검색 트리에 대해 중위 순회가 정렬된 순서로 노드를 방문한다고 가정하자.  

Inductive step: 높이가 h인 임의의 이진 검색 트리 T에 대해 정리가 성립함을 보인다. T의 루트를 r, r의 왼쪽 서브트리를 T_L, 오른쪽 서브트리를 T_R이라 하자. T_L과 T_R은 각각 높이가 h보다 작으므로, 귀납적 가정에 의해 T_L과 T_R 각각의 중위 순회는 정렬된 순서로 노드를 방문한다. (T_L 또는 T_R이 빈 트리인 경우에도 자명하게 성립한다.) BST의 성질에 의해 T_L의 모든 키는 r의 키보다 작고, T_R의 모든 키는 r의 키보다 크다. 따라서 중위 순회 T_L → r → T_R은 전체적으로 정렬된 순서로 노드를 방문한다.

## #09
최대 높이는 n, 최소 높이는 log₂(n+1)

풀이: 각 노드가 자식을 하나씩만 갖는 사향 트리(skewed tree)일 때 높이가 최대가 된다. 이 경우 노드가 n개이면 높이는 n이다. 균형 잡힌 완전 이진 트리에 가까울수록 높이가 최소가 된다. 높이가 h인 이진 트리가 가질 수 있는 최대 노드 수는 $2^{h}$ - 1 개이므로, 최소 높이는 log₂(n+1)이다.
